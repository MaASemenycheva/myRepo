```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
#url="http://www.uniprot.org/uniprot/?query=LRRK2_HUMAN"
url="http://www.uniprot.org/uniprot/?query=ENSP00000309477+AND+organism:9606+gene:ACO1"
```

```python
r=requests.get(url)
```

```python
html_doc=r.text
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
h2_content=soup.main.h2.contents
print(h2_content)
```

```python
main_resource_content = soup.main['resource']
```

```python
print(main_resource_content)
```

```python
searchObj=re.search(r'\/(\w+)$',main_resource_content)
```

```python
print(searchObj.group(1))
```

#Method for Family Download Crawling

```python
new_url="http://www.uniprot.org/uniprot/?query=family:Protein%20Kinase%20A+AND+organism:9606"
```

```python
req=requests.get(new_url)
```

```python
htm_doc=req.text
```

```python
htm_doc
```

```python
doc_handle = BeautifulSoup(htm_doc, 'html.parser')
```

```python
rows=doc_handle.tbody.find_all('tr')
print(len(rows))
```

```python
for row in rows:
    print(row.find_all('td')[2].text)
```
